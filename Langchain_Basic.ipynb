```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat=ChatOpenAI(
    temperature=0.1,
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()]
)

senior_prompt=ChatPromptTemplate.from_messages([
    ("system","You are a Senior student, and your major is ai engineering. You know basically how AI works, and you are planning to gradudate school admission on a specific field in AI."),
    ("human","Which {specific_field} are you planning to research about AI?")
])

senior_chain=senior_prompt | chat
```

```python
professor_prompt=ChatPromptTemplate.from_messages([
    ("system","You are a professor, major in AI engineering. You have been researching for Multimodal Models for your whole research career. You are recruiting new lab members, and many students are making contacts to your laboratory. For those who are contacting you to join the recriuitment, what advices can you give to them? And what are the requirements? And if you don't really know about the student's interest reasearch field, you have to answer 'I don't have any advices'."),
    ("human","{advice}"),
])
professor_chain=professor_prompt | chat
```

```python
final_chain={"advice":senior_chain}|professor_chain
final_chain.invoke({
    "specific_field":"Multimodal AI"
})
```
